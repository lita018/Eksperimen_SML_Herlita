### Deskripsi Dataset
Dataset Heart Disease berisi informasi medis pasien yang digunakan untuk memprediksi apakah seseorang memiliki penyakit jantung atau tidak. Dataset ini sangat umum digunakan dalam penelitian machine learning di bidang kesehatan.

### Informasi Fitur
| Fitur | Tipe | Deskripsi |
|-------|------|-----------|
| age | Numerik | Usia pasien (tahun) |
| sex | Kategorikal | Jenis kelamin (1=Laki-laki, 0=Perempuan) |
| cp | Kategorikal | Tipe nyeri dada (0-3) |
| trestbps | Numerik | Tekanan darah saat istirahat (mm Hg) |
| chol | Numerik | Kolesterol serum (mg/dl) |
| fbs | Kategorikal | Gula darah puasa > 120 mg/dl (1=Ya, 0=Tidak) |
| restecg | Kategorikal | Hasil EKG saat istirahat (0-2) |
| thalach | Numerik | Detak jantung maksimum |
| exang | Kategorikal | Angina akibat olahraga (1=Ya, 0=Tidak) |
| oldpeak | Numerik | Depresi ST akibat olahraga |
| slope | Kategorikal | Kemiringan segmen ST (0-2) |
| ca | Numerik | Jumlah pembuluh darah utama (0-3) |
| thal | Kategorikal | Hasil tes thallium (0-3) |
| **target** | **Biner** | **Label: 1=Ada penyakit jantung, 0=Tidak ada** |

### Tujuan
Membangun model klasifikasi biner untuk memprediksi keberadaan penyakit jantung berdasarkan fitur medis pasien.

# **5. Data Preprocessing**

Pada tahap ini, kita melakukan preprocessing data Heart Disease berdasarkan temuan dari EDA. Tahapan yang dilakukan:
1. Menangani Missing Values (mengisi dengan median)
2. Menghapus Data Duplikat
3. Deteksi dan Penanganan Outlier dengan metode IQR
4. Encoding Data Kategorikal (One-Hot Encoding)
5. Standarisasi Fitur (StandardScaler)
6. Train-Test Split (80:20)

```python
# STEP 1: Handle Missing Values
print('=' * 50)
print('STEP 1: Handle Missing Values')
print('=' * 50)

df_clean = df.copy()
print(f'Missing sebelum: {df_clean.isnull().sum().sum()}')

missing_cols = df_clean.columns[df_clean.isnull().any()].tolist()
for col in missing_cols:
    median_val = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_val)
    print(f'  Kolom "{col}": diisi median = {median_val}')

print(f'Missing sesudah: {df_clean.isnull().sum().sum()}')
print('STEP 1 selesai!')
```

```python
# STEP 2: Hapus Duplikat
print('=' * 50)
print('STEP 2: Hapus Data Duplikat')
print('=' * 50)

before = len(df_clean)
df_clean = df_clean.drop_duplicates().reset_index(drop=True)
print(f'Baris sebelum    : {before}')
print(f'Baris sesudah    : {len(df_clean)}')
print(f'Duplikat dihapus : {before - len(df_clean)} baris')
print('STEP 2 selesai!')
```

```python
# STEP 3: Handle Outlier dengan IQR
print('=' * 50)
print('STEP 3: Handle Outlier (Metode IQR)')
print('=' * 50)

outlier_cols = ['chol', 'trestbps', 'oldpeak']
for col in outlier_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = len(df_clean)
    df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)].reset_index(drop=True)
    print(f'  "{col}": {before - len(df_clean)} outlier dihapus | range [{lower:.1f}, {upper:.1f}]')

print(f'\nShape setelah handling outlier: {df_clean.shape}')
print('STEP 3 selesai!')
```

```python
# STEP 4: One-Hot Encoding
print('=' * 50)
print('STEP 4: One-Hot Encoding')
print('=' * 50)

df_clean['thal'] = df_clean['thal'].fillna(df_clean['thal'].median()).astype(int)

cat_cols = ['cp', 'restecg', 'slope', 'thal']
print(f'Shape sebelum encoding: {df_clean.shape}')
df_encoded = pd.get_dummies(df_clean, columns=cat_cols, drop_first=True)
print(f'Shape setelah encoding : {df_encoded.shape}')
print('STEP 4 selesai!')
```

```python
# STEP 5: Standarisasi (StandardScaler)
print('=' * 50)
print('STEP 5: Standarisasi Fitur')
print('=' * 50)

X = df_encoded.drop('target', axis=1)
y = df_encoded['target']

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print('Statistik setelah scaling (mean mendekati 0, std mendekati 1):')
print(X_scaled.describe().loc[['mean', 'std']].round(3).T.head(5))
print('STEP 5 selesai!')
```

```python
# STEP 6: Train-Test Split
print('=' * 50)
print('STEP 6: Train-Test Split (80:20)')
print('=' * 50)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f'X_train : {X_train.shape}')
print(f'X_test  : {X_test.shape}')
print(f'y_train distribusi: {dict(y_train.value_counts())}')
print(f'y_test  distribusi: {dict(y_test.value_counts())}')
print('STEP 6 selesai!')
```

```python
# STEP 7: Simpan Hasil Preprocessing
print('=' * 50)
print('STEP 7: Simpan Dataset')
print('=' * 50)

os.makedirs('heart_disease_preprocessing', exist_ok=True)

train_df = X_train.copy()
train_df['target'] = y_train.values
test_df = X_test.copy()
test_df['target'] = y_test.values

train_df.to_csv('heart_disease_preprocessing/train.csv', index=False)
test_df.to_csv('heart_disease_preprocessing/test.csv', index=False)

print(f'  train.csv : {train_df.shape}')
print(f'  test.csv  : {test_df.shape}')

print('\n=========================================')
print('SEMUA TAHAPAN PREPROCESSING SELESAI!')
print('=========================================')
print('Ringkasan Preprocessing:')
print('  1. Missing values  -> diisi median')
print('  2. Duplikat        -> diperiksa & dihapus')
print('  3. Outlier         -> IQR method')
print('  4. Encoding        -> One-Hot (cp, restecg, slope, thal)')
print('  5. Scaling         -> StandardScaler')
print('  6. Split           -> 80% train, 20% test (stratified)')
print('  7. Output          -> heart_disease_preprocessing/train.csv & test.csv')
```
